# Do regression analysis to identify the variables that have the strongest relationship with a car's price. 

```python
import statsmodels.api as sm

# Define the predictor variables
X = car[['Year','Engine HP','Engine Cylinders','Number of Doors','highway MPG','city mpg','Popularity','MSRP']]

# Add a constant term to the predictor variables
X = sm.add_constant(X)

# Define the response variable
y = car['MSRP']

# Fit a linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python


# Assuming you have a DataFrame named 'data' with 'market_category' and 'popularity' columns

# Group the data by market category and calculate the average popularity
average_popularity = car.groupby('Market Category')['Popularity'].mean().reset_index()

# Sort the categories by popularity in descending order
average_popularity = average_popularity.sort_values('Popularity', ascending=False)

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(13, 8))

# Create the bar plot for popularity
sns.barplot(x='Market Category', y='Popularity', data=average_popularity, ax=ax1, color='grey')

# Set the y-axis label for the bar plot
ax1.set_ylabel('Popularity')

# Create the line plot for popularity
ax2 = ax1.twinx()
sns.lineplot(x='Market Category', y='Popularity', data=average_popularity, ax=ax2, marker='o', color='black')

# Set the y-axis label for the line plot
ax2.set_ylabel('Popularity')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set the title of the chart
plt.title('Relationship between Market Category and Popularity')

# Display the chart
plt.show()

```

```python


# Assuming you have a DataFrame named 'data' with 'MSRP', 'transmission_type', and 'body_style' columns

# Calculate the average MSRP for each combination of transmission type and body style using groupby and mean
grouped_data = car.groupby(['Transmission Type', 'Vehicle Style'])['MSRP'].mean()

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the scatter plot with different symbols for each body style
for body_style, group in car.groupby('Vehicle Style'):
    ax.scatter(group['MSRP'], group['Transmission Type'], label=body_style)

# Set the x-axis label
ax.set_xlabel('MSRP')

# Set the y-axis label
ax.set_ylabel('Transmission Type')

# Set the title of the chart
plt.title('Relationship between MSRP and Transmission Type')

# Add a legend
plt.legend(title='Body Style',loc='upper right')

# Display the chart
plt.show()
```

```python

# Assuming you have a DataFrame named 'data' with 'MPG', 'body_style', and 'model_year' columns

# Calculate the average MPG for each combination of body style and model year using groupby and mean
grouped_data = car.groupby(['Vehicle Style', 'Make'])['highway MPG'].mean().unstack()

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the line chart for each body style
for body_style, group in grouped_data.groupby('Vehicle Style'):
    group.plot('Vehicle Style', body_style, ax=ax, label=body_style)

# Set the x-axis label
ax.set_xlabel('Model Year')

# Set the y-axis label
ax.set_ylabel('MPG')

# Set the title of the chart
plt.title('Trend of Fuel Efficiency (MPG) over Time for Each Body Style')

# Add a legend
plt.legend()

# Display the chart
plt.show()

```
